```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/kaggle/input/second-hand-car-price-prediction/cars.csv")
```

```python
df.head(2)
```

```python
df.describe()
```

```python
# Histogram of car prices
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
# Pie chart of fuel type distribution
fuel_counts = df['Fuel_Type'].value_counts()
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Fuel Type Distribution')
plt.show()
```

```python
# Bar plot of transmission type distribution
transmission_counts = df['Transmission'].value_counts()
transmission_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.title('Transmission Type Distribution')
plt.show()
```

```python
# Count plot of owner type distribution
sns.countplot(x='Owner_Type', data=df, palette='pastel')
plt.xlabel('Owner Type')
plt.ylabel('Count')
plt.title('Owner Type Distribution')
plt.show()
```

```python
# Scatter plot of kilometers driven vs price
plt.scatter(df['Kilometers_Driven'], df['Price'], alpha=0.5)
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.title('Scatter Plot of Kilometers Driven vs Price')
plt.show()
```

```python
# Box plot of price vs fuel type
sns.boxplot(x='Fuel_Type', y='Price', data=df, palette='Set3')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('Box Plot of Price vs Fuel Type')
plt.show()
```

```python
# Box plot of price vs transmission type
sns.boxplot(x='Transmission', y='Price', data=df, palette='Set2')
plt.xlabel('Transmission Type')
plt.ylabel('Price')
plt.title('Box Plot of Price vs Transmission Type')
plt.show()
```

```python
# Bar plot of car brand distribution (top 10 brands)
top_brands = df['Brand'].value_counts().head(10)
top_brands.plot(kind='bar', color='salmon')
plt.xlabel('Car Brand')
plt.ylabel('Count')
plt.title('Top 10 Car Brands Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of car model distribution (top 10 models)
top_models = df['Model'].value_counts().head(10)
top_models.plot(kind='bar', color='lightblue')
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.title('Top 10 Car Models Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot of price vs owner type
sns.boxplot(x='Owner_Type', y='Price', data=df, palette='Set1')
plt.xlabel('Owner Type')
plt.ylabel('Price')
plt.title('Box Plot of Price vs Owner Type')
plt.show()
```

```python
# Histogram of car age (year)
current_year = 2024  # Assuming the current year is 2024
df['Car_Age'] = current_year - df['Year']
plt.hist(df['Car_Age'], bins=15, color='lightgreen', edgecolor='black')
plt.xlabel('Car Age (Years)')
plt.ylabel('Frequency')
plt.title('Distribution of Car Age')
plt.show()
```

```python
# Pairplot of numerical variables with hue for fuel type
sns.pairplot(df[['Price', 'Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats', 'Fuel_Type']], hue='Fuel_Type', palette='husl')
plt.title('Pairplot of Numerical Variables with Hue for Fuel Type')
plt.show()
```
